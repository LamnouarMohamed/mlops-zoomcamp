```python
import pandas as pd
import pickle
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_parquet('https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2023-01.parquet')
```

```python
df.head()
```

```python
df['duration'] = df.tpep_dropoff_datetime - df.tpep_pickup_datetime
```

```python
df.head()
```

```python
df.duration = df.duration.apply(lambda td: td.total_seconds() / 60)
```

```python
df.head()
```

```python
df.shape
```

```python
'duration' in df.columns
```

```python
df.duration.std()
```

```python
df = df[(df.duration >= 1) & (df.duration <= 60)]
```

```python
df.shape
```

```python
3009173 / 3066766
```

```python
categorical = ['PULocationID', 'DOLocationID']
df[categorical].dtypes
```

```python
df[categorical] = df[categorical].astype(str)
df[categorical].dtypes
```

```python
df.head()
```

```python
train_dicts = df[categorical].to_dict(orient='records')
train_dicts
```

```python
dv = DictVectorizer()
X_train = dv.fit_transform(train_dicts)
```

```python
X_train.shape
```

```python
target = 'duration'
y_train = df[target].values

lr = LinearRegression()
lr.fit(X_train, y_train)

y_pred = lr.predict(X_train)
```

```python
from sklearn.metrics import root_mean_squared_error
```

```python
# for solve the previous problem you can use below code
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(y_train, y_pred)
```

```python
import numpy as np
rmse = np.sqrt(mse)
```

```python
rmse
```

```python
# for validation:

df_validation = pd.read_parquet('https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2023-02.parquet')
df_validation['duration'] = df_validation.tpep_dropoff_datetime - df_validation.tpep_pickup_datetime
df_validation.duration = df_validation.duration.apply(lambda td: td.total_seconds() / 60)
df_validation = df_validation[(df_validation.duration >= 1) & (df_validation.duration <= 60)]
```

```python
categorical = ['PULocationID', 'DOLocationID']
df_validation[categorical] = df_validation[categorical].astype(str)
val_dicts = df_validation[categorical].to_dict(orient='records')
X_val = dv.transform(val_dicts)
target = 'duration'
y_val = df_validation[target].values
y_val_pred = lr.predict(X_val)
mse = mean_squared_error(y_val, y_val_pred)
rmse = np.sqrt(mse)
rmse
```
